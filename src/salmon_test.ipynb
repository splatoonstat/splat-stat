```python
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from packages.japanize import japanize
import packages.db as db
import packages.definitions as d

lobby = d.SalmonLobby.NORMAL
date_from = dt.date(2023, 5, 1)
date_to = dt.date(2023, 5, 7)
works = db.load_works(lobby=lobby, date_from=date_from, date_to=date_to)
```

```python
works["stage"].value_counts()
```

```python
sns.set_theme()
japanize()

sns.displot(
    data=works,
    y="title-after",
    height=4,
    aspect=2,
)
plt.show()
```

```python
sns.set_theme()
japanize()

sns.displot(
    data=works,
    x="hazard-level",
    aspect=2,
)
plt.show()
```

```python
sns.set_theme()
japanize()

sns.relplot(
    data=works,
    x="hazard-level",
    y="job-point",
    hue="cleared-waves",
    palette="flare",
    s=7.5,
)
plt.show()
```
