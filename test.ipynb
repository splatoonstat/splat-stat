```python
import datetime as dt
import packages.db as db
import packages.definitions as d

battles = db.load_battles(lobby=d.Lobby.XMATCH, date_from=dt.date(2023, 3, 1), date_to=dt.date(2023, 3, 31))
battles["mode"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
sns.displot(
    data=battles,
    x="power",
    hue="mode",
    multiple="stack",
)
plt.show()
```
